## import library yang akan digunakan

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
```

```python
df = pd.read_csv('labeling data.csv')
```

## untuk menampilkan jumlah data pada masing2 label sentimen

```python
df_disini=pd.read_csv('labeling data.csv')
df_disini['sentimen'].value_counts()
```

## untuk menampilkan data apa saja yang memiliki label positif, negatif dan netral

```python
#melihat sentimen positif, negatif dan netral
j=1
sortedDF = df.sort_values(by=['sentimen'])
for i in range (0,sortedDF.shape[0]):
    if (sortedDF['sentimen'][i]=='negatif'):
        print (str(j)+')'+sortedDF['tweet'][i])
        print()
        j=j+1
```

## diagram sentimen 

```python
labels=list(df_disini['sentimen'].value_counts().index)
sizes=list(df_disini['sentimen'].value_counts().values)
explode=(0, 0.1, 0)

fig1,ax1=plt.subplots()
ax1.pie(sizes,explode=explode, labels=labels, autopct='%1.1f%%', 
        shadow=True, startangle=10)
ax1.axis('equal')
plt.show()
```

```python
words=''
for txt in df_disini['tweet']:
    words +=str(txt)
blob=TextBlob(words)
blob.word_counts
```

## wordcloud

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

kata=WordCloud(width=500, height=300,background_color='white').generate(words)
plt.figure(figsize=(12,8))
plt.imshow(kata)
plt.title('vaksin Covid-19', size=22)
plt.axis('off')
plt.show()
```
